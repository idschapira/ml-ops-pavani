```python
# leitura da tabela com as variáveis candidatas a features
df_feat_candidate = spark.table("workspace.default.bz_hr_attr_features")
```

```python
from pyspark.sql.functions import when, col

#Converte as variáveis categóricas em numéricas
df_feat_candidate = df_feat_candidate.withColumn(
    "OverTime", when(col("OverTime") == "Yes", 1).otherwise(0)
).withColumn(
    "Attrition", when(col("Attrition") == "Yes", 1).otherwise(0)
)
```

```python
df_feat_candidate = df_feat_candidate.withColumn(
    "MaritalStatusNum",
    when(col("MaritalStatus") == "Married", 3)
    .when(col("MaritalStatus") == "Single", 2)
    .when(col("MaritalStatus") == "Divorced", 1)
    .otherwise(0)
)

display(df_feat_candidate, limit=10)
```

```python
# O que acontece com o startdadization?

from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd

# Supondo que 'df' é o seu DataFrame e 'features_numericas' são as colunas a serem normalizadas
features_categoricas = [
    'MaritalStatusNum',
    'JobLevel',
    'JobInvolvement',
    'YearsAtCompany',
    'MonthlyIncome',
    'StockOptionLevel',
    'JobSatisfaction',
    'EnvironmentSatisfaction',
    'OverTime',
    'WorkLifeBalance'
]

#clona para fazer os experimentos com padronizacao de valores
df_feat_std = df_feat_candidate.toPandas().copy()

# --- Usando StandardScaler ---
scaler_standard = StandardScaler()
df_feat_std[features_categoricas] = scaler_standard.fit_transform(df_feat_std[features_categoricas])
print("DataFrame após Padronização (StandardScaler):\n", df_feat_std)

display(df_feat_std, limit=10)
```

```python
#clona para fazer os experimentos com padronizacao de valores
df_feat_minmax = df_feat_candidate.toPandas().copy()

# --- Usando MinMaxScaler ---
scaler_minmax = MinMaxScaler()
df_feat_minmax[features_categoricas] = scaler_minmax.fit_transform(df_feat_minmax[features_categoricas])

display(df_feat_minmax, limit=10)
```

```python
spark.conf.set("spark.driver.extraJavaOptions", "-Dorg.apache.spark.ml.feature.StringIndexer=true")
```

```python
# Importar as bibliotecas necessárias
from pyspark.sql import SparkSession
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Definir as features categóricas e numéricas
# features_categoricas = ["col1", "col2"]  # Substitua pelos nomes das suas colunas categóricas
numerical_features = ["num_col1", "num_col2"]  # Substitua pelos nomes das suas colunas numéricas

# Pipelines para processamento de features
stages = []

# Processamento de features categóricas: StringIndexer + OneHotEncoder
for col in features_categoricas:
    indexer = StringIndexer(inputCol=col, outputCol=col + "_indexed")
    encoder = OneHotEncoder(inputCol=col + "_indexed", outputCol=col + "_encoded")
    stages += [indexer, encoder]

# Processamento de features numéricas: StandardScaler (opcional, mas recomendado)
# Note: StandardScaler espera uma única coluna Vector, então precisamos de um VectorAssembler antes dele
# assembler_numerical = VectorAssembler(inputCols=numerical_features, outputCol="numerical_features_vector")
# scaler = StandardScaler(inputCol="numerical_features_vector", outputCol="scaled_numerical_features")
# stages += [assembler_numerical, scaler]

# Juntar todas as features em uma única coluna 'features'
encoded_output_cols = [f + "_encoded" for f in features_categoricas]
final_features = encoded_output_cols + ["scaled_numerical_features"]  # Use "numerical_features_vector" se não usar scaler

assembler = VectorAssembler(inputCols=final_features, outputCol="features")
stages.append(assembler)

# Criar o Pipeline
pipeline = Pipeline(stages=stages)

# Ajustar o pipeline aos dados para transformar as features
pipeline_model = pipeline.fit(df_feat_std)
df_transformed = pipeline_model.transform(df_feat_std)

# Selecionar as colunas finais para o modelo
model_data = df_transformed.select("features", "Attrition")

# Dividir os dados em treino e teste
(trainingData, testData) = model_data.randomSplit([0.8, 0.2], seed=42)

# Treinar o modelo de Regressão Logística
lr = LogisticRegression(featuresCol='features', labelCol='Attrition')
lr_model = lr.fit(trainingData)

# Fazer previsões no conjunto de teste
predictions = lr_model.transform(testData)

# Mostrar algumas previsões
predictions.select("features", "label", "prediction", "probability").show(truncate=False)

# Avaliar o modelo
evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="Attrition", metricName="areaUnderROC")
auc = evaluator.evaluate(predictions)
print(f"Area Under ROC (AUC) = {auc}")
```
